```python
import numpy as np
import pandas as pd
import os
df = pd.read_csv(os.getcwd() + '/Laliga.csv')
df
```

```python
df = df.replace(['-'],['0'])
df
```

```python
for debut in df.Debut:
    if debut.find("-") != -1:
        splitted = debut.split("-")
        value = splitted[0][:2]+splitted[1]
    else:
        value = debut
    
    if(value > '1930' and value < '1980'):
        print((df.loc[df['Debut'] == debut]['Team']).values, debut)
    
#create tabular format output
```

```python
topteam = pd.to_numeric(df['Points']).nlargest(5)
for tp in topteam.values:
    print(df.loc[pd.to_numeric(df['Points']) == tp]['Team'].values, tp)
```

```python
def Goal_diff_count(GoalsFor, GoalsAgainst):
    return (GoalsFor - GoalsAgainst)

for i, row in df.iterrows():
    print(row['Team'], Goal_diff_count(int(row['GoalsFor']), int(row['GoalsAgainst'])))
```

```python
GoalDiff = []
for i, row in df.iterrows():
    GoalDiff.append(Goal_diff_count(int(row['GoalsFor']), int(row['GoalsAgainst'])))

df['diff'] = GoalDiff 
print(df.loc[df['diff'] == df['diff'].max()]['Team'].values, df['diff'].max())
print(df.loc[df['diff'] == df['diff'].min()]['Team'].values, df['diff'].min())
df = df.drop('diff',axis=1) 
```

```python
def winning_percent (GamesWon, GamesPlayed):
    if(GamesPlayed == 0):
        return 0
    else:
        return((GamesWon/GamesPlayed)*100)
    

winPerc = []
for i, row in df.iterrows():
    winPerc.append(winning_percent(int(row['GamesWon']), int(row['GamesPlayed'])))

df['Winning Percent'] = winPerc 
df
```

```python
highwin = df['Winning Percent'].nlargest(5)
for hw in highwin.values:
    print(df.loc[df['Winning Percent'] == hw]['Team'].values, hw)
```
